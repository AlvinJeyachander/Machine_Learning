```python
import turicreate #imported the turicreate library
```

#Load house sales dataset

```python
sales = turicreate.SFrame('home_data.sframe')
```

```python
sales
```

```python
#Exploring data for housing sales:
```

```python
sales.show()
```

```python
turicreate.show(sales[1:5000]['sqft_living'],sales[1:5000]['price'])
```

```python
#create a simple regression model of Sq_ft Living to price:
```

```python
training_set , test_set = sales.random_split(0.8,seed=0)
```

```python
#Building the regression model:
```

```python
sqft_model = turicreate.linear_regression.create(training_set,target = 'price',features = ['sqft_living'])
```

```python
#Evaluate Simple model:
```

```python
print (test_set['price'].mean())
```

```python
print(sqft_model.evaluate(test_set))
```

```python
#visualization of how predictions look like :
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(test_set['sqft_living'],test_set['price'],'.',
         test_set['sqft_living'],sqft_model.predict(test_set),'-')
```

```python
sqft_model.coefficients
```
